Vamos a definir una clase llamada **`UsuarioNetflix`**.

Esta clase tendrá nuestra información de usuario, así como un todas las películas que hemos visto, y las puntuación de cada pelicula.

- Define una clase llamada **`UsuarioNetflix`** que tenga los siguientes atributos:
1. **`username`** que será el nombre del usuario.
2. **`movies`** que será un diccionario:
    - Las llaves de este diccionario serán los titulos de las peliculas.
    - Los valores de cada llave será un diccionario que tendrá las siguientes llaves:
        - La puntuación de cada pelicula dada por el usuario.
        - La fecha en la en la que el usuario vio la pelicula (**`Dia-Mes-Año`**).
        - Genero.
    - Debe ser inicializado vacio.
    
- La clase tendrá los siguientes métodos:
1. **`ver_pelicula()`** que tome los siguientes parámetros:
    - **`titulo`**: el titulo de la pelicula.
    - **`puntuacion`**: la puntuacion del usuario.
    - **`genero`**: el género de la película.
    - **`fecha`**: la fecha de cuando el usuario vio la pelicula, esta fecha debe estar en formato **`Dia-Mes-Año`**. Transforma esta fecha a un objeto **`datetime`**. Haz que por defecto la fecha sea el día de hoy.
    
    Éste método deberá hacer lo siguiente:
    - Agrege la información al diccionario **`movies`** tal que:
```python
{titulo : {"puntuacion" : puntuacion, "fecha" : fecha, "genero" : genero}}
```
    - El método no retorna nada.
    
    
2. **`display_peliculas()`** que no tome ningún parámetro:
    - Muestre por pantalla la lista de peliculas que el usuario a visto, seguido de su puntuación, la fecha y el género.
        - **`Titulo: Sipderman - Puntuación: 5 - Fecha: 01-01-2022 - Género: Acción`**.
    - La función no retorna nada.
    
    
3. **`mas_gustados()`** que no tome ningún parámetro:
    - Encuentre cual es la puntuación más alta dada por el usuario. Guardalo en la variable **`max_puntuacion`**.
    - Retorne una lista con las peliculas que tengan la puntuación más alta (**`max_puntuacion`**). Llama a la lista **`peliculas`**.
    
    
4. **`menos_gustados()`** que no tome ningún parámetro:
    - Encuentre cual es la puntuación más baja dada por el usuario. Guardalo en la variable **`min_puntuacion`**.
    - Retorne una lista con las peliculas que tengan la puntuación más baja (**`min_puntuacion`**). Llama a la lista **`peliculas`**.
    
    
5. **`info_genero()`** que no tome ningún parámetro:
    - Imprima por pantalla la siguiente información:
        - Género más visto.
        - Género menos visto.
        - Género con la puntuación más alta.
        - Género con la puntuación más baja.
        - Todos los géneros vistos por el usuario.
        - El método no retorna nada.
        
        
6. **`info_fechas()`** que no time ningún parámetro:
    - Imprima por pantalla la siguiente información:
        - Mes cuando se vió más peliculas.
        - Mes cuando se vió menos peliculas.
        - Año cuando se vió más peliculas.
        - Año cuando se vió menos peliculas.
        - Dia de la semana cuando se ve más peliculas (El resultado deberá ser un día de la semana).
        - Dia de la semana cuando se ve menos peliculas (El resultado deberá ser un día de la semana).
        
        
6. **`exportar_info()`** que no tome ningún parámetro:
    - Guarde en un archivo **`.pkl`** el diccionario **`movies`**. El archivo debe llamarse **`usuario_netflix_info.pkl`**.
    - El método no retorna nada.

````python
from datetime import datetime
from collections import Counter
from collections import defaultdict
import pickle


class UsuarioNetflix:
    username = ""
    movies = {}
    max_puntuacion = 0
    min_puntuacion = 10
    
    def __init__(self,username):
        self.username = username
        self.max_puntuacion = 0
        self.min_puntuacion = 10
    
    def ver_pelicula(self, titulo,puntuacion,genero,fecha):
        # movies ->  {titulo : {"puntuacion" : puntuacion, "fecha" : fecha, "genero" : genero}}
        self.movies[titulo] = {
            "puntuacion": puntuacion,
            "fecha": fecha,
            "genero": genero
        }

        
        
         ################ ESTO ES POR MESES ################

    def display_peliculas(self):        
        for movie in self.movies:
            print(movie, '->', self.movies[movie]['puntuacion'], self.movies[movie]['fecha'], self.movies[movie]['genero'])
        
         ################ MAS GUSTADO ################
            
    def mas_gustados(self):
        for movie in self.movies:
            if (self.movies[movie]['puntuacion'] > self.max_puntuacion):
                self.max_puntuacion = self.movies[movie]['puntuacion']
        print('Maxima puntuacion', self.max_puntuacion)
        print('Mas gustadas:')
        for movie in self.movies:
            if (self.movies[movie]['puntuacion'] == self.max_puntuacion):
                print(movie, '->', self.movies[movie]['puntuacion'], self.movies[movie]['fecha'], self.movies[movie]['genero'])
 
                ################ MENOS GUSTADO ################
                
    def menos_gustados(self):
        for movie in self.movies:
            if (self.movies[movie]['puntuacion'] < self.min_puntuacion):
                self.min_puntuacion = self.movies[movie]['puntuacion']
        print('Minima puntuacion', self.min_puntuacion)
        print('Menos gustadas:')
        for movie in self.movies:
            if (self.movies[movie]['puntuacion'] == self.min_puntuacion):
                print(movie, '->', self.movies[movie]['puntuacion'], self.movies[movie]['fecha'], self.movies[movie]['genero'])                
                
                 ################ INFORMACION GENEROS ################
        
    def info_genero(self):
#1 Filtrado
        groups = defaultdict(list)       
        for obj in self.movies:
            groups[self.movies[obj]['genero']].append(obj)
        new_list = groups.values()
        dadp = list(new_list)
        sfr = list(groups)
        print("El genero mas visto es:"+str(sfr[0])+" y sus pelis son:"+str(dadp[0])+"\n")
 #Tengo un problema si lo paso a listas solo puedo pasar el primero o el ultimo
# 2 Filtrado
        print("El genero menos visto es:"+str(sfr[-1])+" y sus pelis son:"+str(dadp[-1])+"\n")
        groups = defaultdict(list)
        for obj in self.movies:
            groups[self.movies[obj]['genero']].append(self.movies[obj]["puntuacion"])
        print (groups)
        for key, value in groups.items():
            groups[key] = sum(value)/len(value)
# Lista de pelis y medias QUE NO SE ME OLVIDE COSTEARLO A LISTAS 
        print(list(groups))
# lista de generos  QUE NO SE ME OLVIDE COSTEARLO A LISTAS 
        print(list(groups.keys()))

    
    def info_fechas(self):
        
        ################ ESTO ES POR MESES ################
        Meses = list()
        for peli,valor in self.movies.items():
            Meses.append(valor['fecha'].month)
        Meses = set(Meses)
        print(Meses)
        print("-"*100)    
        dicc_Meses = dict()
        for Mes in Meses:
            dicc_Meses[Mes] = list()
            for peli,valor in self.movies.items():
                if valor['fecha'].month == Mes:
                    dicc_Meses[Mes].append(valor)

        # Ordenamos la lista por tamaño
        #print("Antes de ordenar")
        #print(dicc_Meses)
        Meses_Ordenados_Tamano = sorted(dicc_Meses.items(), key=lambda kv: (len(kv[1]), kv[0]))
        #print("Despues de ordenar")
        #print(Meses_Ordenados_Tamano)

        # El mes que menos peliculas se han visto
        print("Mes que menos pelis se han visto:", Meses_Ordenados_Tamano[0])
        print("Mes es el ", Meses_Ordenados_Tamano[0][0])
        print("Numero de pelis ese mes:", len(Meses_Ordenados_Tamano[0][1]))
        
        # El mes que mas peliculas se han visto
        print("Mes que mas pelis se han visto:", Meses_Ordenados_Tamano[len(Meses_Ordenados_Tamano) - 1])
        print("Mes es el ", Meses_Ordenados_Tamano[len(Meses_Ordenados_Tamano) - 1][0])
        print("Numero de pelis ese mes:", len(Meses_Ordenados_Tamano[len(Meses_Ordenados_Tamano) - 1][1]))
        
        ################ ESTO ES POR AÑOS ################
        
        Anios = list()
        for peli,valor in self.movies.items():
            Anios.append(valor['fecha'].year)
        Anios = set(Anios)
        print("-"*100)
        print(Anios)
        print("-"*100)    
        dicc_Anios = dict()
        for Anio in Anios:
            dicc_Anios[Anio] = list()
            for peli,valor in self.movies.items():
                if valor['fecha'].year == Anio:
                    dicc_Anios[Anio].append(valor)

        # Ordenamos la lista por tamaño
        #print("Antes de ordenar")
        #print(dicc_Anios)
        Anios_Ordenados_Tamano = sorted(dicc_Anios.items(), key=lambda kv: (len(kv[1]), kv[0]))
        #print("Despues de ordenar")
        #print(Anios_Ordenados_Tamano)

        # El mes que menos peliculas se han visto
        print("Año que menos pelis se han visto:", Anios_Ordenados_Tamano[0])
        print("Año es el ", Anios_Ordenados_Tamano[0][0])
        print("Numero de pelis ese Año:", len(Anios_Ordenados_Tamano[0][1]))
        
        # El mes que mas peliculas se han visto
        print("Año que mas pelis se han visto:", Anios_Ordenados_Tamano[len(Anios_Ordenados_Tamano) - 1])
        print("Año es el ", Anios_Ordenados_Tamano[len(Anios_Ordenados_Tamano) - 1][0])
        print("Numero de pelis ese Año:", len(Anios_Ordenados_Tamano[len(Anios_Ordenados_Tamano) - 1][1]))
        
        ################ ESTO ES POR DIAS DE LA SEMANA ################
        DiasSemana = list()
        for peli,valor in self.movies.items():
            DiasSemana.append(valor['fecha'].weekday())
        DiasSemana = set(DiasSemana)
        print("-"*100)
        print(DiasSemana)
        print("-"*100)    
        dicc_DiasSemana = dict()
        for DiaSemana in DiasSemana:
            dicc_DiasSemana[DiaSemana] = list()
            for peli,valor in self.movies.items():
                if valor['fecha'].weekday() == DiaSemana:
                    dicc_DiasSemana[DiaSemana].append(valor)

        # Ordenamos la lista por tamaño
        #print("Antes de ordenar")
        #print(dicc_DiasSemana)
        DiasSemana_Ordenados_Tamano = sorted(dicc_DiasSemana.items(), key=lambda kv: (len(kv[1]), kv[0]))
        #print("Despues de ordenar")
        #print(DiasSemana_Ordenados_Tamano)
        
        def dayNameFromWeekday(weekday):
            if weekday == 0:
                return "Lunes"
            if weekday == 1:
                return "Martes"
            if weekday == 2:
                return "Miercoles"
            if weekday == 3:
                return "Jueves"
            if weekday == 4:
                return "Viernes"
            if weekday == 5:
                return "Sabado"
            if weekday == 6:
                return "Domingo"
        
        # El dia semana que menos peliculas se han visto
        print("DiaSemana que menos pelis se han visto:", dayNameFromWeekday(DiasSemana_Ordenados_Tamano[0][0]))
        print("DiaSemana es el ", DiasSemana_Ordenados_Tamano[0][0])
        print("Numero de pelis ese dia semana:", len(DiasSemana_Ordenados_Tamano[0][1]))
        
        # El dia semana que mas peliculas se han visto
        print("DiaSemana que mas pelis se han visto:", dayNameFromWeekday(DiasSemana_Ordenados_Tamano[len(DiasSemana_Ordenados_Tamano) - 1][0]))
        print("DiaSemana es el ", DiasSemana_Ordenados_Tamano[len(DiasSemana_Ordenados_Tamano) - 1][0])
        print("Numero de pelis ese dia semana:", len(DiasSemana_Ordenados_Tamano[len(DiasSemana_Ordenados_Tamano) - 1][1]))
        
        
    def guardar(self):
        with open('usuario_netflix_info.pkl', 'wb') as esc:
            pickle.dump(self.movies, esc)
            esc.close()
        
usuario = UsuarioNetflix("Jorge")
usuario.ver_pelicula("Avatar", 7, "Ciencia Ficciónn", datetime(2018,8,15))
usuario.ver_pelicula("Harry Potter", 5, "Ciencia Ficciónn", datetime(2015,2,10))
usuario.ver_pelicula("Otra de ciencia Ficcion", 3, "Ciencia Ficciónn", datetime(2015,2,10))
usuario.ver_pelicula("Karate Kid", 4, "Comedia", datetime(2016,9,1))
usuario.ver_pelicula("El señor de los anillos", 9, "Fantasía", datetime(2012,9,1))
usuario.ver_pelicula("Otra peli en la 9", 9, "Fantasía", datetime(2012,9,1))

print("-"*100)
usuario.info_fechas()
print("-"*100)
usuario.info_genero()
print("-"*100)
usuario.menos_gustados()
print("-"*100)
usuario.mas_gustados()
print("-"*100)
usuario.guardar()
````
